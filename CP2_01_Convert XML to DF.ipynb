
PseudoCode to Convert XML to Pandas Dataframe
1. Parse the Posts.xml file using xml.dom.minidom
2. Test whether the parsing worked by getting a sample of items from the file
3. Create a header row from the items in each row of the xml
4. Create an empty list to append data to
5. Write a for loop to extract all items from each row of the xml (make sure to write an N/A for those fields that are empty)
6. Create a dataframe with the extracted data and headers
7. QC dataframe


```
'''
Import all modules that are needed
'''
import xml.dom.minidom
import pandas as pd
import pickle
```

```
'''
1.0 Parse Posts.xml file
'''
posts = xml.dom.minidom.parse('/content/drive/My Drive/Capstone2/Data/Posts.xml')
```

```
'''
2a. Test whether the parsing worked by getting a sample of items from the file
'''
print(posts.nodeName)
print(posts.firstChild.tagName)
print(posts.getElementsByTagName("row"))
rows = posts.getElementsByTagName("row")
```

```
'''
2b. Using a for loop, test whether I can access all the questions in the body
'''
for item in rows:
  print(item.getAttribute("Tags"))
```

```
'''
3. Create a header row from the items in the rows of the XML 
'''
df_cols = ['Id','PostTypeId','CreationDate','Score','ViewCount','Body','OwnerUserId','LastActivityDate','Title','Tags','AnswerCount','CommentCount','FavoriteCount','ClosedDate','ContentLicense']
```

```
'''
4. Create an empty list to append data to
5. Write a for loop to extract all items from each row of the xml
6. Create a dataframe with the extracted data and headers
'''
extracted_data = []

for item in rows: 
    Id = item.getAttribute("Id")
    PostTypeId = item.getAttribute("PostTypeId") 
    CreationDate = item.getAttribute("CreationDate") 
    Score = item.getAttribute("Score") 
    ViewCount = item.getAttribute("ViewCount") 
    Body = item.getAttribute("Body") 
    OwnerUserId = item.getAttribute("OwnerUserId") 
    LastActivityDate = item.getAttribute("LastActivityDate") 
    Title = item.getAttribute("Title") 
    Tags = item.getAttribute("Tags") 
    AnswerCount = item.getAttribute("AnswerCount") 
    CommentCount = item.getAttribute("CommentCount") 
    FavoriteCount = item.getAttribute("FavoriteCount") 
    ClosedDate = item.getAttribute("ClosedDate") 
    ContentLicense = item.getAttribute("ContentLicense") 
    
    extracted_data.append({'Id':Id, 'PostTypeId':PostTypeId, 'CreationDate':CreationDate, 'Score':Score, 'ViewCount':ViewCount, 'Body':Body, 'OwnerUserId':OwnerUserId,\
                           'LastActivityDate':LastActivityDate,'Title':Title, 'Tags':Tags, 'AnswerCount':AnswerCount, 'CommentCount':CommentCount, 'FavoriteCount':FavoriteCount,\
                           'ClosedDate': ClosedDate, 'ContentLicense':ContentLicense})

posts_df = pd.DataFrame(extracted_data, columns = df_cols)
```

```
'''
ALL CODE COMBINED
'''
posts = xml.dom.minidom.parse('/content/drive/My Drive/Capstone2/Data/Posts.xml')
rows = posts.getElementsByTagName("row")

df_cols = ['Id','PostTypeId','CreationDate','Score','ViewCount','Body','OwnerUserId','LastActivityDate','Title','Tags','AnswerCount','CommentCount','FavoriteCount',\
           'ClosedDate','ContentLicense']

extracted_data = []

for item in rows: 
    Id = item.getAttribute("Id")
    PostTypeId = item.getAttribute("PostTypeId")
    CreationDate = item.getAttribute("CreationDate")
    Score = item.getAttribute("Score")
    ViewCount = item.getAttribute("ViewCount")
    Body = item.getAttribute("Body")
    OwnerUserId = item.getAttribute("OwnerUserId")
    LastActivityDate = item.getAttribute("LastActivityDate")
    Title = item.getAttribute("Title")
    Tags = item.getAttribute("Tags")
    AnswerCount = item.getAttribute("AnswerCount")
    CommentCount = item.getAttribute("CommentCount")
    FavoriteCount = item.getAttribute("FavoriteCount")
    ClosedDate = item.getAttribute("ClosedDate")
    ContentLicense = item.getAttribute("ContentLicense")
    
    extracted_data.append({'Id':Id, 'PostTypeId':PostTypeId, 'CreationDate':CreationDate, 'Score':Score, 'ViewCount':ViewCount, 'Body':Body, 'OwnerUserId':OwnerUserId,\
                           'LastActivityDate':LastActivityDate,'Title':Title, 'Tags':Tags, 'AnswerCount':AnswerCount, 'CommentCount':CommentCount, 'FavoriteCount':FavoriteCount,\
                           'ClosedDate': ClosedDate, 'ContentLicense':ContentLicense})

posts_df = pd.DataFrame(extracted_data, columns = df_cols)
```
